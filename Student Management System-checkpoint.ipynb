```python
import mysql.connector
from tkinter import *
from tkinter import messagebox
from tkinter import ttk

win=Tk()
win.title('STUDENT MANAGEMENT SYSTEM')

win.minsize(width=800,height=600)
win.configure(bg='skyblue')

def add():
    name=n1.get()
    mob_no=m1.get()
    dept=d1.get()
    email_id=e1.get()
    
    if(name=='' or mob_no=='' or dept=='' or email_id==''):
        messagebox.showinfo('info','All fields are compulsary')
    else:
        conn=mysql.connector.connect(user="root",password="localhost@123",
                             host="localhost",auth_plugin='mysql_native_password',database="projectdb")


        qur="INSERT INTO student(name,mob_no,dept,email_id) VALUES('%s','%s','%s','%s')"%(name,mob_no,dept,email_id)
        cursor=conn.cursor()
        cursor.execute(qur)
        conn.commit()
        conn.close()
        messagebox.showinfo("info","Successfully Added!")
        n1.set('')
        m1.set('')
        d1.set('')
        e1.set('')
    
    
def show():
    treev.delete(*treev.get_children())

    conn=mysql.connector.connect(user="root",password="localhost@123",
                             host="localhost",auth_plugin='mysql_native_password',database="projectdb")
    qur="SELECT * FROM student"
    cursor=conn.cursor()
    cursor.execute(qur)
    lst=cursor.fetchall()
    for i in lst:
        treev.insert("",'end',values=(i[0],i[1],i[2],i[3]))
    conn.close()
        
    
def result():
    root=Tk()
    root.title('Result Analysis')
    root.minsize(width=800,height=600)
    def disp():
        name=e7.get()
        if name=='':
            messagebox.showinfo('info','This field is compulsary')
        else:
            conn=mysql.connector.connect(user="root",password="localhost@123",
                             host="localhost",auth_plugin='mysql_native_password',database="projectdb")

            qur='select mob_no from student where name="%s"'%(name)
            cursor=conn.cursor()
            cursor.execute(qur)
            tup1=cursor.fetchone()
            l8.config(text=tup1[0]+'%')
        
    l7=Label(root,text='Name of Student',width=20,relief='ridge',
             bd=5,fg='black',bg='white',font=('times new roman',12,'bold'))
    l7.place(x=100,y=80)
    
    e7=Entry(root,width=40,relief='ridge',
             bd=5,fg='black',bg='white',font=('times new roman',12,'bold'))
    e7.place(x=300,y=80)
    
    b7=Button(root,text='Show Result',relief='ridge',bd=5,width=20,command=disp)
    b7.place(x=300,y=120)
    
    l8=Label(root,text='Marks here',height=5,width=20,relief='ridge',
             bd=5,fg='black',bg='white',font=('times new roman',12,'bold'))
    l8.place(x=100,y=160)
    
    root.mainloop()

def delete():
    del_name = tt.get()
    
    if del_name == '':
        messagebox.showinfo("info","Enter name to be deleted.")
    else:
        conn=mysql.connector.connect(user="root",password="localhost@123",
                             host="localhost",auth_plugin='mysql_native_password',database="projectdb")
        #messagebox.askquestion("askquestion", "Are you sure?")
          
        qur="DELETE FROM student where name='%s'"%(del_name)
        try:
            mycur=conn.cursor()
            cursor=conn.cursor()
            cursor.execute(qur)
            conn.commit()
        except:
            conn.rollback
        conn.close()
        messagebox.showinfo("info","Successfully Deleted!")
        tt.set('')

    

def select():
    name= updata.get()
    if name == '':
        messagebox.showinfo("info","Enter name to be deleted.")
    else:
        conn=mysql.connector.connect(user="root",password="localhost@123",
                             host="localhost",auth_plugin='mysql_native_password',database="projectdb")
        qur=' SELECT * FROM student WHERE name="%s"'%(name)
        cursor=conn.cursor()
        cursor.execute(qur)
        lst = cursor.fetchall()
        for i in lst:
            ent1.insert(0,i[0])
            ent2.insert(0,i[1])
            ent3.insert(0,i[2])
            ent4.insert(0,i[3])
        conn.close()
        updata.set('')
   
    
def update():
    name=n1.get()
    mob_no=m1.get()
    dept=d1.get()
    email_id=e1.get()
    
    if(name=='' or mob_no=='' or dept=='' or email_id==''):
        messagebox.showinfo('info','All fields are compulsary')
    else:
        conn=mysql.connector.connect(user="root",password="localhost@123",
                             host="localhost",auth_plugin='mysql_native_password',database="projectdb")


        qur="UPDATE student SET mob_no='%s',dept='%s',email_id='%s' where name='%s'"%(mob_no,dept,email_id,name)
        cursor=conn.cursor()
        cursor.execute(qur)
        conn.commit()
        conn.close()
        messagebox.showinfo("info","Successfully Updated!")
        n1.set('')
        m1.set('')
        d1.set('')
        e1.set('')
    
       
# clearing treeview data
def clear():
    treev.delete(*treev.get_children())

# Main label for Project Name
lbl=Label(win,text="STUDENT MANAGEMENT SYSTEM",
      relief='ridge',bd=5,fg='black',bg='white',font=('times new roman',16,'bold'))
lbl.place(x=268,y=40)


# variable for entry
n1=StringVar()
m1=StringVar()
d1=StringVar()
e1=StringVar()

#name label and entry
lbl1=Label(win,text='Name',width=20,relief='ridge',bd=5,fg='black',bg='white',font=('times new roman',12,'bold'))
lbl1.place(x=190,y=100)
ent1=Entry(win,bg='white',fg='black',width=40,bd=5,textvariable=n1,font=('times new roman',12))
ent1.place(x=385,y=100)

# mobile number label and entry
lbl2=Label(win,text='Mobile Number',width=20,relief='ridge',bd=5,fg='black',bg='white',font=('times new roman',12,'bold'))
lbl2.place(x=190,y=140)
ent2=Entry(win,bg='white',fg='black',width=40,bd=5,textvariable=m1,font=('times new roman',12))
ent2.place(x=385,y=140)


# department label and entry
lbl3=Label(win,text='Department',width=20,relief='ridge',bd=5,fg='black',bg='white',font=('times new roman',12,'bold'))
lbl3.place(x=190,y=180)
ent3=Entry(win,bg='white',fg='black',width=40,bd=5,textvariable=d1,font=('times new roman',12))
ent3.place(x=385,y=180)

# email id label and entry
lbl4=Label(win,text='Email ID',width=20,relief='ridge',bd=5,fg='black',bg='white',font=('times new roman',12,'bold'))
lbl4.place(x=190,y=220)
ent4=Entry(win,bg='white',fg='black',width=40,bd=5,textvariable=e1,font=('times new roman',12))
ent4.place(x=385,y=220)

# Buttons : Show,Add,Exit,Result
showbtn=Button(win,text='SHOW',width=20,command=show).place(x=200,y=320)
addbtn=Button(win,text='ADD',width=20,command=add).place(x=500,y=320)
exitbtn=Button(win,text='EXIT',width=20,command=win.destroy).place(x=200,y=360)
relbtn=Button(win,text='RESULT',command=result,width=20).place(x=500,y=360)

# Delete label entry button
tt=StringVar()
lbl5=Label(win,text='Delete by Name >>',width=20,relief='ridge',bd=5,fg='black',bg='white',font=('times new roman',10,'bold'))
lbl5.place(x=190,y=400)
ent5=Entry(win,bg='white',fg='black',width=40,bd=5,textvariable=tt,font=('times new roman',12))
ent5.place(x=385,y=400)
delbtn=Button(win,text=' Delete',width=20,command=delete).place(x=300,y=440)

# Update label entry button
lbl6=Label(win,text='Update by Name >>',width=20,
           relief='ridge',bd=5,fg='black',bg='white',
           font=('times new roman',10,'bold'))
lbl6.place(x=190,y=500)
updata=StringVar()
ent6=Entry(win,bg='white',fg='black',width=40,bd=5,
           textvariable=updata,font=('times new roman',12))
ent6.place(x=385,y=500)
upbtn=Button(win,text=' Update',width=20,command=update).place(x=450,y=540)
selbtn=Button(win,text=' Select',width=20,command=select).place(x=300,y=540)

# Tree View
treev = ttk.Treeview(win, selectmode ='browse',height=20)
treev.place(x=900,y=100,width=450)

treev["columns"] = ("1", "2", "3","4")
treev['show'] = 'headings'
treev.column("1", width = 90, anchor ='c')
treev.column("2", width = 90, anchor ='se')
treev.column("3", width = 90, anchor ='se')
treev.column("4", width = 90, anchor ='se')

treev.heading("1", text ="Name")
treev.heading("2", text ="Mobile No")
treev.heading("3", text ="Dept")
treev.heading("4", text ="EmailID")


# clear the tree view
cleartbtn=Button(win,text=' Clear',width=20,command=clear).place(x=1080,y=560)


win.mainloop()
```
